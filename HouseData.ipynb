```python
df=df.select_dtypes(include=['float64','int64']).copy()
```

```python
df.dropna(inplace=True)
```

```python
sns.jointplot(df['LotArea'],df['SalePrice'])
```

```python
df= df[df['LotArea']<50000]
```

```python
sns.jointplot(df['YearBuilt'],df['SalePrice'])
```

```python
df= df[df['SalePrice']<500000]
```

```python
df= df[df['YearBuilt']>1900]
```

```python
sns.boxplot(x='OverallCond',y='SalePrice',data=df)
```

```python
X= df.drop('SalePrice',axis=1)
y=df['SalePrice']
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predic=lm.predict(X_test)
```

```python
from sklearn import metrics
metrics.mean_absolute_error(y_test,predic)
```

```python
t=metrics.mean_squared_error(y_test,predic)
t
```
