# Excersise on various Visualization Techniques

The iris dataset contains measurements for 150 iris flowers from three different species.


The three classes in the Iris dataset:
Iris-setosa (n=50)
Iris-versicolor (n=50)
Iris-virginica (n=50)
The four features of the Iris dataset:
sepal length in cm
sepal width in cm
petal length in cm
petal width in cm


Get the data
*Use seaborn to get the iris data by using: iris = sns.load_dataset('iris') *



```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

Create a kde plot of sepal_length versus sepal width for setosa species of flower.

```python
meta=iris[iris['species']=='setosa']
sns.kdeplot(meta['sepal_width'],meta['sepal_length'],cmap='plasma',shade=True,shade_lowest=False)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('ticks')
tips['tips_pct'] = tips['tip']/tips['total_bill']*100
a = sns.FacetGrid(tips, col = 'time', row = 'sex',margin_titles = True)
a = a.map(plt.hist,'tips_pct')
```

```python
plt.figure(figsize=(5.5,6))
sns.boxplot('day','total_bill',data=tips,hue='sex',palette='seismic')
sns.despine()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
fig,ax=plt.subplots()
x=np.linspace(0,10,100)
ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x),'--')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
x=np.linspace(0,10,100)
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")
axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.8,0.8])
axes.spines['right'].set_color('none')
axes.spines['top'].set_color('none')
axes.xaxis.set_ticks_position('bottom')
axes.spines['bottom'].set_position(('data',0))
axes.yaxis.set_ticks_position('left')
axes.spines['left'].set_position(('data',0))
x = np.linspace(-0.75, 1, 100)
axes.plot(x, x**3);
```
